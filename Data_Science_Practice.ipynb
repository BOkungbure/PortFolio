```python
df_pop = pd.read_csv('population_total.csv')
```

```python
df_pop.dropna(inplace=True)
```

```python
df_pivot = df_pop.pivot(
    index='year',
    columns='country',
    values='population')
```

```python
df_pivot = df_pivot[['United States','India','China','Brazil','Indonesia']]
```

```python
df_pivot.plot(
    kind='line',
    xlabel='Year',
    ylabel='Population',
    title='Population (1955-2020)',
    figsize=(10,4)
    )

plt.savefig('line_chart.png')
plt.show()
```

```python
df_pivot_2020 = df_pivot[df_pivot.index.isin([2020])]
```

```python
df_pivot_2020
```

```python
df_pivot_2020 = df_pivot_2020.T
```

```python
df_pivot_2020.plot(
    kind='bar',
    color='green',
    xlabel='Year',
    ylabel='Population',
    title='Population (2020)',
    figsize=(10,4)
    )
```

```python
df_pivot_years = df_pivot[df_pivot.index.isin([1980,1990,2000,2010,2020])]
```

```python
df_pivot_years.plot(
    kind='bar',
    xlabel='Year',
    ylabel='Population',
    title='Population (1980-2020)',
    figsize=(10,4))
```
